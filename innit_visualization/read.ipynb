```python
# Hàm phân loại thời gian
def classify_time(time):
    if time >= pd.to_datetime('00:00:00', format='%H:%M:%S').time() and time < pd.to_datetime('06:00:00', format='%H:%M:%S').time():
        return 'Early_Morning'
    elif time >= pd.to_datetime('06:00:00', format='%H:%M:%S').time() and time < pd.to_datetime('12:00:00', format='%H:%M:%S').time():
        return 'Morning'
    elif time >= pd.to_datetime('12:00:00', format='%H:%M:%S').time() and time < pd.to_datetime('17:00:00', format='%H:%M:%S').time():
        return 'Afternoon'
    elif time >= pd.to_datetime('17:00:00', format='%H:%M:%S').time() and time < pd.to_datetime('20:00:00', format='%H:%M:%S').time():
        return 'Evening'
    elif time >= pd.to_datetime('20:00:00', format='%H:%M:%S').time() and time < pd.to_datetime('23:00:00', format='%H:%M:%S').time():
        return 'Night'
    else:
        return 'Late_Night'

# Áp dụng hàm classify_time cho cột dep_time để tạo cột mới
```

```python
df_march_test['departure_time'] = df_march_test['dep_time'].apply(classify_time)
```
