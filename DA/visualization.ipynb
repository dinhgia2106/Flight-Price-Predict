```python
import pandas as pd 

df = pd.read_csv('../data/november_full_data_encoded.csv')
df.head()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(df['price'], kde=True, bins=30, color='skyblue')
plt.title("Price Distribution of Flight Tickets")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.title('Airline Distribution')
airline_counts = df['airline_code'].value_counts()
plt.pie(airline_counts, labels=airline_counts.index, autopct='%1.1f%%', colors = ["#C6E2FF","#B9D3EE" ,"#9FB6CD","#EEE9E9","#6C7B8B","#CDC9C9"])
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='airline_code', y='price', data=df)
plt.title("Price Distribution by Airline")
plt.xlabel("Airline")
plt.ylabel("Price")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='departure_time', y='price', data=df)
plt.title("Impact of Departure Time on Ticket Prices")
plt.xlabel("Departure Time")
plt.ylabel("Price")
plt.xticks(rotation=45)
plt.show()
```

```python
airline_avg_price = df.groupby('airline')['price'].mean()


airline_avg_price.plot(kind='bar', color='skyblue', edgecolor='black')


plt.title('Price vs Airline')
plt.xlabel('Airline')
plt.ylabel('Average Price')


plt.show()
```

```python
print(df.value_counts('airline_code'))
print(df.value_counts('Routine'))
```

```python
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

scatter = ax.scatter(df['airline_code'], 
                    df['Routine'],
                    df['price'],
                    c=df['price'], 
                    cmap='viridis',
                    alpha=0.6)

plt.colorbar(scatter)

ax.set_xlabel('Airline')
ax.set_ylabel('Routine') 
ax.set_zlabel('Price (VND)')

plt.title('Relationship between Price, Airline and Routine')

plt.show()
```

```python
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

scatter = ax.scatter(df['airline_code'], 
                    df['dep_time'],
                    df['price'],
                    c=df['price'], 
                    cmap='viridis',
                    alpha=0.6)

plt.colorbar(scatter)

ax.set_xlabel('Airline')
ax.set_ylabel('Departure Time') 
ax.set_zlabel('Price (VND)')

plt.title('Relationship between Price, Airline and Departure Time')

plt.show()
```
